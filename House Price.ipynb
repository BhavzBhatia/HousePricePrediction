```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
hp = pd.read_csv("HousePrice.csv")
hp
```

```python
col = pd.get_dummies(hp.town)
col
```

```python
new_hp = pd.concat([hp,col], axis = 'columns')
new_hp
```

```python
final = new_hp.drop(["town", "west windsor"], axis = 1)
```

```python
final
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
x = final.drop("price", axis = 1)
y = final.price
```

```python
reg.fit(x,y)
reg.predict([[2800,0,1]])    # price of 2800 sqft house in Robinsville
```

```python
reg.predict([[3400,0,0]]) #price of a 3400sqft house in West Windsor
```
